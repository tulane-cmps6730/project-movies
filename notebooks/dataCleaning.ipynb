```python
# List of columns you want to keep
columns_to_keep = ['GameType', 'Location', 'Date', 'Time', 'WinningTeam', 'Quarter',
                   'SecLeft', 'AwayTeam', 'AwayPlay', 'AwayScore', 'HomeScore','HomeTeam', 'HomePlay']
# Assuming `df` is your pandas DataFrame
categorical_columns = ['GameType', 'Location', 'WinningTeam', 'AwayTeam', 'HomeTeam']

for col in categorical_columns:
    df[col] = df[col].astype('category')

# For Date and Time columns, you might want to convert them to datetime objects if not already
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M %p').dt.time  # Assuming Time is in the format "8:00 PM"

# Confirm the dtypes after conversion
print(df.dtypes)

# Select only the columns you want to keep
df_filtered = df.loc[:, columns_to_keep]
```

Dropping all rows where there is no text for home or away play

```python
df_filtered = df_filtered.dropna(subset=['AwayPlay', 'HomePlay'], how='all')
```

Organizing so that features makeup first and second half.

```python


# Sort the DataFrame to ensure it is ordered by game and then by quarter
df_filtered = df_filtered.sort_values(by=['Date', 'AwayTeam', 'HomeTeam', 'Quarter'])

# Add a game identifier if one doesn't exist
# Ensure all columns involved in concatenation are converted to strings
df_filtered['GameID'] = df_filtered['Date'].astype(str) + '_' + df_filtered['AwayTeam'].astype(str) + '_' + df_filtered['HomeTeam'].astype(str)
```

```python
import pandas as pd

# Assuming your dataframe is named 'df'

# Replace 'NaN' string with actual NaN
df_filtered['AwayPlay'].replace('NaN', pd.NA, inplace=True)
df_filtered['HomePlay'].replace('NaN', pd.NA, inplace=True)

# Create a new column 'MergedPlay' that combines 'AwayPlay' and 'HomePlay'
df_filtered['MergedPlay'] = df_filtered['AwayPlay'].combine_first(df['HomePlay'])

# If you want to fill the 'nan' with an empty string instead, use this line:
# df['MergedPlay'] = df['AwayPlay'].combine_first(df['HomePlay']).fillna('')

```

FOllowing will give you a df with all plays in quarter as features and final play as label

```python
import pandas as pd

def consolidate_game_text_with_label_by_quarter(df):
    # Initialize an empty list to hold the consolidated data
    consolidated_data = []
    
    # Group the DataFrame by 'GameID' and 'Quarter', then iterate over each group
    for (gameID, quarter), group in df.groupby(['GameID', 'Quarter']):
        # Sort the group by 'SecLeft' in descending order
        group = group.sort_values(by='SecLeft', ascending=False)
        
        # Check if the group has more than one row to separate input and label
        if len(group) > 1:
            # Separate the last row in time (first row after sorting) as the label, and the rest as the gameText
            label_row = group.iloc[0]  # The first row after sorting is the last in time
            input_rows = group.iloc[1:]  # Exclude the first row for the gameText
        else:
            # If there's only one play in a quarter, use it as both input and label
            label_row = group.iloc[0]
            input_rows = group.iloc[:0]  # Empty DataFrame for input
        
        # Concatenate 'SecLeft' and 'MergedPlay' for each row in the input_rows
        gameText = ':'.join(f"{row['SecLeft']} {row['MergedPlay']}" for index, row in input_rows.iterrows())
        
        # Prepare the label text from the label row
        label_text = f"{label_row['SecLeft']} {label_row['MergedPlay']}"
        
        # Append the consolidated data as a tuple (gameID, quarter, gameText, label_text)
        consolidated_data.append((gameID, quarter, gameText, label_text))
    
    # Create a new DataFrame with the consolidated data
    output_df = pd.DataFrame(consolidated_data, columns=['GameID', 'Quarter', 'GameText', 'Label'])
    
    return output_df

# Assuming 'df' is your DataFrame
# Example usage:
new_df_with_labels_by_quarter = consolidate_game_text_with_label_by_quarter(df_filtered)
new_df_with_labels_by_quarter.head(1)
```

This uses previous 5 rows as features, and 5th as label

```python
import pandas as pd

def consolidate_game_text_with_label_by_quarter(df):
    # Initialize an empty list to hold the consolidated data
    consolidated_data = []
    
    # Group the DataFrame by 'GameID' and 'Quarter', then iterate over each group
    for (gameID, quarter), group in df.groupby(['GameID', 'Quarter']):
        # Sort the group by 'SecLeft' in descending order
        group = group.sort_values(by='SecLeft', ascending=False)
        
        # Iterate over the group to create sequences of 5 GameTexts with the next as the label
        for i in range(len(group) - 5):
            # Get 5 GameTexts as the features
            gameTexts = ':'.join(f"{row['SecLeft']} {row['MergedPlay']}" for index, row in group.iloc[i:i+5].iterrows())
            # Get the following GameText as the label
            label_row = group.iloc[i+5]
            label_text = f"{label_row['SecLeft']} {label_row['MergedPlay']}"
            
            # Append the consolidated data as a tuple (gameID, quarter, gameTexts, label_text)
            consolidated_data.append((gameID, quarter, gameTexts, label_text))
    
    # Create a new DataFrame with the consolidated data
    output_df = pd.DataFrame(consolidated_data, columns=['GameID', 'Quarter', 'GameText', 'Label'])
    
    return output_df

# Assuming 'df' is your DataFrame
# Example usage:
new_df_with_labels_by_quarter = consolidate_game_text_with_label_by_quarter(df_filtered)
new_df_with_labels_by_quarter.head()
```
