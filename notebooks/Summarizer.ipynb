```python
from spacy.matcher import Matcher
import spacy
def add_custom_patterns(nlp):
    matcher = Matcher(nlp.vocab)
    # Pattern for 'Initial. Lastname'
    # Using "SHAPE": "X." may match the initial and period but needs proper shaping for last name.
    pattern = [{"TEXT": {"REGEX": "^[A-Z]\\.$"}}, {"IS_ALPHA": True, "LENGTH": {">": 1}}]  # Regex for Initial. and normal token for Lastname
    matcher.add("PLAYER_NAME", [pattern])
    return matcher

# Integrate this function within the SpaCy pipeline
nlp = spacy.load("en_core_web_sm")
matcher = add_custom_patterns(nlp)

def extract_entities_with_matcher(text):
    doc = nlp(text)
    matches = matcher(doc)
    players = set(doc[start:end].text for _, start, end in matches)  # Use a set to avoid duplicates
    return list(players)

# Example usage

players = extract_entities_with_matcher(text)
print(players)
```

```python
import pandas as pd
import spacy
from collections import defaultdict

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    players = [ent.text for ent in doc.ents if ent.label_ == "PERSON"]
    return players

def quantify_actions(text):
    actions = defaultdict(lambda: defaultdict(int))
    sentences = text.split(". ")
    
    for sentence in sentences:
        players = extract_entities(sentence)
        player = players[0] if players else None
        
        if player:
            if "misses" in sentence:
                actions[player]['misses'] += 1
            elif "makes" in sentence:
                actions[player]['makes'] += 1
                if "dunk" in sentence:
                    actions[player]['dunks'] += 1
                if "assist" in sentence:
                    assist_by = sentence.split("assist by")[-1].strip()
                    actions[assist_by]['assists'] += 1
            elif "rebound" in sentence:
                type_of_rebound = "Defensive rebound" if "Defensive rebound" in sentence else "Offensive rebound"
                actions[player][type_of_rebound] += 1
            elif "foul" in sentence:
                actions[player]['fouls'] += 1
            elif "free throw" in sentence:
                outcome = "made FT" if "makes" in sentence else "missed FT"
                actions[player][outcome] += 1
            elif "Turnover" in sentence:
                actions[player]['turnovers'] += 1
            elif "steal" in sentence:
                actions[player]['steals'] += 1

    return actions


def generate_summary(actions):
    category_leaders = defaultdict(lambda: ('', 0))  # Dictionary to track the leader for each category
    
    # Calculate leaders in each category
    for player, categories in actions.items():
        for category, count in categories.items():
            if count > category_leaders[category][1]:
                category_leaders[category] = (player, count)

    summaries = []
    # Create summaries for significant categories
    significant_categories = ['misses', 'makes', 'Defensive rebound', 'Offensive rebound', 
                              'fouls', 'turnovers', 'steals', 'assists', 'made FT', 'missed FT', 'dunks']
    for category in significant_categories:
        player, count = category_leaders[category]
        if player and player != "Unknown Player":  # Ignore 'Unknown Player' and categories not mentioned
            action_phrase = "occurrences" if category not in ['misses', 'makes'] else "times"
            summaries.append(f"{player} led {category} with {count} {action_phrase}.")

    return " ".join(summaries)



# Assume df is your DataFrame and 'CleanGameText' is your column
df = pd.read_csv("Data/Actual_Quarter.csv")
text = str(df["CleanGameText"].iloc[0:4].tolist())  # Assuming taking the first entry

actions_dict = quantify_actions(text)
summary = generate_summary(actions_dict)
print("Game Summary:", summary)
```
